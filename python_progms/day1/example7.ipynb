```python
# check prime or not prime 
n = int (input("enter a number to check prime or not prime "))
ch= False
if n>1 :
    for i in range (2 ,n):
        if (n % i ) == 0 :
            ch= True
            break
    if ch == True:
        print( n , "is not a prime number")
    else:
        print(n , "is a prime number")
```

```python
# alternate

n=int(input())

if n<= 1 :
    print("not prime")
else :
    for i in range (2, n):
        if n%i == 0 :
            print("not prime")
            break
    else:
        print("prime")
```
