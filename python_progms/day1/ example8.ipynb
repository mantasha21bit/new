```python
l1=[2 , 5, [6,7] ,8 ,21,42]
l2=l1[:]
print("ref of l1",id(l1))
print("ref of l2",id(l2))
print(l1)
print(l2)

print("ref of l1[2]",id(l1[2]))
print("ref of l2[2]",id(l2[2]))
l2[2][1]=44
print(l1)
print(l2)
```

```python
from copy import deepcopy
l1=[2 , 5, [6,7] ,8 ,21,42]
l2=deepcopy(l1)
print("ref of l1",id(l1))
print("ref of l2",id(l2))
print(l1)
print(l2)

print("ref of l1[2]",id(l1[2]))
print("ref of l2[2]",id(l2[2]))
l2[2][1]=44
print(l1)
print(l2)
```
